# Defining pca 

```python
class pca(object):
    def __init__(self,x,k):
        self.x=x
        self.k=k

    def feat_mean(self):
        '''creating a feature mean function'''
        mew=np.zeros(self.x.shape[1])
        mew
        for i in range(self.x.shape[1]):
            mew[i]=sum(self.x[:,i])/self.x.shape[0]
        return mew
    
    def shift_matrix(self): #creating a shift matrix
        mew=self.feat_mean()
        return(self.x-mew)

    def vc_matrix(self):
        n=self.x.shape[0]
        sigma=self.shift_matrix().T.dot(self.shift_matrix())/(n-1)
        return sigma
    def eig(self):
        sigma=self.vc_matrix()
        eigenval,eigenvec=np.linalg.eig(sigma)
        return eigenvec
    def pc(self):
        k_c=self.eig()[:,0:self.k]
        return k_c
    def projected_data(self):
        pcs=self.pc()
        z=self.shift_matrix()
        return(z.dot(pcs))

```
