# DBSCAN Clustering

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import math 

centers = [[1, 1], [-1, -1], [1, -1]]
x, y = make_blobs(n_samples=100, centers=centers,random_state=2,cluster_std=0.2)
fig,ax = plt.subplots(figsize=(10,5))
ax.set_title("Actual Data showing 3 clusters")
ax.scatter(x[:,0],x[:,1],c=y,s=50,cmap = 'inferno')
data=pd.DataFrame({'points':x.tolist(),'cluster':0}) #creating the intial dataset
data.head()
```

# Creating a class 

```python
class dbscan(object):
    def __init__(self,x,eps,m):
        self.x=x
        self.eps=eps
        self.m=m
    def nbhd_x(self,n): #defining the neighbourhood function
        nbhd=[]
        for i in range(len(self.x)):
                d=math.sqrt(sum([(a - b) ** 2 for a, b in zip(self.x[n], self.x[i])])) #calculating the distance
                if d!=0 and d<self.eps: 
                    nbhd.append(i) #appending the index of neighbouring point
        return(nbhd)
    def iterate_cluster(self): #iterarions
        cluster=1 #initialising the cluster
        for i in range(len(self.x)):
             if data.iloc[i,1]==0:
                a=self.nbhd_x(i)
                if len(a)>=self.m:
                    data.iloc[i,1]=cluster
                    for j in a:
                        data.iloc[j,1]=cluster #assigning parent cluster to neighbour points
                    cluster=cluster+1
                else:
                    data.iloc[i,1]=-1 #declaring outliers
        return(data)
```

```python
db=dbscan(x,0.5,23)
db.iterate_cluster()
```
